```python
import torch
import torch.nn as nn
```

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, padding=1)
        self.pool = nn.MaxPool2d(2,2)
        self.conv2 = nn.Conv2d(16,32,3,padding=1)
        self.fc1 = nn.Linear(32*7*7,128)
        self.fc2 = nn.Linear(128,10)
    def forward(self,x):
        a = self.pool(torch.relu(self.conv1(x)))
        b = self.pool(torch.relu(self.conv2(a)))
        c = b.view(b.size(0),-1)
        d = torch.relu(self.fc1(c))
        e = self.fc2(d)
        return e
```

```python
img = torch.rand(1,1,28,28)
cnn = CNN()
out = cnn(img)
print('The size after the first convolution :: ', cnn.conv1.weight.shape)
print('The size after the second convolution :: ', cnn.conv2.weight.shape);
print('The size after the first fully connected layer :: ', cnn.fc1.weight.shape)
print('The size after the second fully connected layer :: ', cnn.fc2.weight.shape)
print('the final output size',out.shape)
```
